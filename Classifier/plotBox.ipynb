```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib import text
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import cv2
```

```python
# csv PATH
CSV_PATH = '/content/cutsPred.csv'

# original image dir
ORIGINAL_IMG_PATH = "/content/originalImg/"
```

```python
# import csv file that are need to predict
terns_columns = ['filename', 'xmin', 'ymin', 'xmax', 'ymax', 'predictLabel']
terns_to_label = pd.read_csv(CSV_PATH, header=0, names=terns_columns)
terns_to_label['filename'] = terns_to_label['filename'].astype(str)

# get the original file name for each crop imgae into a new column
terns_to_label['originFile'] = terns_to_label['filename'].apply(lambda x: x.split("_")[2])
```

```python
terns_to_label
```

```python
def plotBoxOnOriginalImage(imgName, terns_to_label):
    '''
    imgName: the file name of original image
    terns_to_label: dataframe of imgs that we want to predict
    '''

    # get the token to specify which original image we try to plot
    target = imgName.split("_")[2][:-4]

    # read the image
    image = cv2.imread(ORIGINAL_IMG_PATH + imgName)

    targetDataFrame = terns_to_label[terns_to_label["originFile"] == target]
    targetDataFrame = targetDataFrame[["xmin", "ymin", "xmax", "ymax", "predictLabel"]]

    for row in targetDataFrame.iterrows():
        xmin, ymin, xmax, ymax, label = row[1]
        top_left = (xmin, ymax)
        bottom_right = (xmax, ymin)
        if label == "ROTE":
          colour = (0, 0, 255)
        elif label == "SANE":
          colour = (0, 255, 0)
        image = cv2.rectangle(image, top_left, bottom_right, colour, thickness = 3)
    return cv2.imwrite(ORIGINAL_IMG_PATH + "BOX_" + imgName, image)
```

```python
imgList = ["DJI_20220521112941_0179.JPG", 
          "DJI_20220521113138_0279.JPG", 
          "DJI_20220521113419_0418.JPG", 
          "DJI_20220521113836_0640.JPG",
           ]
```

```python
for imgName in imgList:
  plotBoxOnOriginalImage(imgName, terns_to_label)
```
